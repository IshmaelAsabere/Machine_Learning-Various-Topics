#### Iterating 

**df.iteritems()**  | Iterates w.r.t a column and its corresponding values as Key-Pair<br>
**df.iterrows()**   | Iterate over each row as (key, value) pair<br>
**df.itertuples()** | Returns an iterator yeilding a named tuple for each row

```python
# (Key, Value) pair iterates over the column labels as the key and its content as the values associated with the key
import pandas as pd
import numpy as np

df =pd.DataFrame(np.random.rand(5, 4), columns=['col1', 'col2', 'col3', 'col4'])
print(df)

print()

for key, value in df.iteritems():
    print(key, value)
```

```python
# df.iterrows(): Iterate over each row as (key, value) pair
import pandas as pd
import numpy as np
df =pd.DataFrame(np.random.rand(5, 4),
                 columns=['col1', 'col2', 'col3', 'col4'])
print(df)
print()
for key, value in df.iterrows():
    print(key, value)
```

```python
# df.itertuples(): Returns an iterator yeilding a named tuple for each row
import pandas as pd
import numpy as np
df =pd.DataFrame(np.random.rand(5, 4), 
                 columns=['col1', 'col2', 'col3', 'col4'])
for row in df.itertuples():
    print(row)
```

**df1.join(df2,on=col1, how='inner')** | SQL-style join the columns in df1 with the columns on df2 where the rows for col have identical values. 'how' can be one of 'left', 'right', 'outer', 'inner'<br>

```python
# The data we will use for the joins
import pandas

# Creating the World Champions DataFrame
world_champions={'Team':['India','Australia','West Indies','Pakistan','Sri Lanka'],
                 'ICC_rank':[2,3,7,8,4],
                 'World_champions_Year':[2011,2015,1979,1992,1996],
                 'Points':[874,787,753,673,855]}

# Creating the Chokers DataFrame
chokers={'Team':['South Africa','New Zealand','Zimbabwe'],
         'ICC_rank':[1,5,9],
         'Points':[895,764,656]}

df1=pandas.DataFrame(world_champions)
df2=pandas.DataFrame(chokers)

print(df1)
print()
print(df2)
```

```python
# LeftJoin
print(pandas.merge(df1,df2,on='Team',how='left' ))
```

```python
# Right Join
print(pandas.merge(df1,df2,on='Team',how='right' ))
```

```python
# Outer Join
print(pandas.merge(df1,df2,on='Team',how='outer' ))
```

```python
# Inner Join
print(pandas.merge(df1, df2, on ='Team', how = 'inner'))
```
